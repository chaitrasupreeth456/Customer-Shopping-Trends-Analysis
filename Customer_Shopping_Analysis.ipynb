```python
# Loading the dataset using pandas
import pandas as pd
df = pd.read_csv('customer_shopping_data.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
#checking whether any duplicates
df.duplicated().all()
```

```python
# Checking if missing data or null values are present in the dataset
df.isnull().sum()
```

```python
# Adding a new column with a flag if review rating is missing
df['review_rating_was_missing'] = df['Review Rating'].isna().astype(int)
```

```python
df.head(10)
```

```python
# Filling missing values in 'Review Rating' column with the median rating of the product category
df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Converting column names to snake case
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column 'age_group' based on 4 age groups
bins =[0,12,19,59,100]
labels=['Child','Teenager','Adult','Senior']
df['age_group']=pd.cut(df['age'],bins=bins,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
# create new column based on "frequency_of_purchases"

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(15)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
#check whether column 'discount_applied' and 'promo_code_used' are same?
if (df['discount_applied'] == df['promo_code_used']).all():
    print("All matched")
```

```python
#Dropping a dupliacte column
df=df.drop('promo_code_used',axis=1)
```
